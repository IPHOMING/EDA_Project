```python
#Create a line graph using seaborn library, showing the relationship between "Academic Year" and "Average Annual Salary (HK$'000)" from the DataFrame 'df'. 

linegraph = sns.lineplot(x="Academic Year", y="Average Annual Salary (HK$'000)",linewidth=4, hue='Broad Academic Programme Category', data=df, err_style=None)

linegraph.set_xticklabels(linegraph.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

```python

#Generate a line graph using seaborn library to visualize the relationship between "Academic Year" and "Average Annual Salary (HK$'000)" from the DataFrame 'df'. 

linegraph = sns.lineplot(x="Academic Year", y="Average Annual Salary (HK$'000)", hue='Level of Study', data=df, err_style=None)

linegraph.set_xticklabels(linegraph.get_xticklabels(), rotation=40, ha="right")

plt.tight_layout()

plt.show()
```

```python
#copy df name as df2
df2=df
```

```python
#filter rows where 'Academic Year' == '2021/22'

df2=df2[df2['Academic Year']=='2021/22']

df2
```

```python
#This code groups the data by 'Broad Academic Programme Category', calculates the mean of 'Average Annual Salary (HK$'000)', and sorts the results in descending order.

df_mean=df2.groupby('Broad Academic Programme Category')['Average Annual Salary (HK$\'000)'].mean()
df_mean= pd.DataFrame(df_mean)
df_mean.reset_index(inplace=True)
df_sorted = df_mean.sort_values("Average Annual Salary (HK$'000)", ascending=False)
df_sorted
```

```python
#This code creates a bar plot to show the relationship between Broad Academic Programme Category and average salaries.

barplot=sns.barplot(x = 'Broad Academic Programme Category',
            y = "Average Annual Salary (HK$'000)",
            data = df_sorted)

barplot.set_xticklabels(barplot.get_xticklabels(), rotation=40, ha="right")
barplot.legend(fontsize='7')
plt.tight_layout()
plt.show()
```

```python
#This code generates a bar plot with grouped bars, representing average salaries by Academic Programme categories and levels of study.

barplot=sns.barplot(x = 'Broad Academic Programme Category',
            y = "Average Annual Salary (HK$'000)",
            hue = 'Level of Study',
            data = df)

barplot.set_xticklabels(barplot.get_xticklabels(), rotation=40, ha="right")
barplot.legend(fontsize='7')
plt.tight_layout()
plt.show()


```
