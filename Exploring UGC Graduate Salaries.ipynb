```python
linegraph = sns.lineplot(x="Academic Year" ,y="Average Annual Salary (HK$'000)", hue='Broad Academic Programme Category', data=df,err_style=None)
```

```python
# Create the line plot
linegraph = sns.lineplot(x="Academic Year", y="Average Annual Salary (HK$'000)", hue='Broad Academic Programme Category', data=df, err_style=None)

linegraph.set_xticklabels(linegraph.get_xticklabels(), rotation=40, ha="right")

plt.tight_layout()

plt.show()
```

```python
linegraph = sns.lineplot(x="Academic Year", y="Average Annual Salary (HK$'000)", hue='Level of Study', data=df, err_style=None)

linegraph.set_xticklabels(linegraph.get_xticklabels(), rotation=40, ha="right")

plt.tight_layout()

plt.show()
```
